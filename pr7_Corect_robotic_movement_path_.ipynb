```python
import matplotlib.pyplot as plt

class Robot:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.direction = 'N'  # N, E, S, W
        self.path = [(self.x, self.y)]

    def move_forward(self):
        if self.direction == 'N':
            self.y += 1
        elif self.direction == 'E':
            self.x += 1
        elif self.direction == 'S':
            self.y -= 1
        elif self.direction == 'W':
            self.x -= 1
        self.path.append((self.x, self.y))

    def turn_left(self):
        dirs = ['N', 'W', 'S', 'E']
        self.direction = dirs[(dirs.index(self.direction) + 1) % 4]

    def turn_right(self):
        dirs = ['N', 'E', 'S', 'W']
        self.direction = dirs[(dirs.index(self.direction) + 1) % 4]

    def show_path(self):
        xs, ys = zip(*self.path)
        plt.plot(xs, ys, marker='o')
        plt.title('Robot Path')
        plt.grid(True)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.axis('equal')
        plt.show()

# Instantiate the robot
robot = Robot()

# Command sequence
robot.move_forward()
robot.move_forward()
robot.turn_right()
robot.move_forward()
robot.move_forward()
robot.turn_left()
robot.move_forward()

# Show movement path
robot.show_path()
```
