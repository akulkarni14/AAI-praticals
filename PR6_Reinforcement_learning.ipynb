```python
import random

# Game setup
boxes = [' '] * 9
HUMAN = 'X'
COMPUTER = 'O'
first_player = HUMAN
turn = 1

winning_combos = [
    [0, 1, 2], [3, 4, 5], [6, 7, 8],
    [0, 3, 6], [1, 4, 7], [2, 5, 8],
    [0, 4, 8], [2, 4, 6],
]

def print_board(initial=False):
    board = [str(i + 1) if initial else boxes[i] for i in range(9)]
    print(f'''
 {board[0]} | {board[1]} | {board[2]}
-----------
 {board[3]} | {board[4]} | {board[5]}
-----------
 {board[6]} | {board[7]} | {board[8]}
''')

def take_turn(player, turn):
    while True:
        if player == COMPUTER:
            box = get_computer_move()
        else:
            try:
                box = int(input(f'Player {player}, choose a box (1-9): ')) - 1
            except ValueError:
                print('Invalid input. Enter a number between 1 and 9.\n')
                continue

        if box < 0 or box > 8:
            print('That number is out of range. Try again.\n')
            continue

        if boxes[box] == ' ':
            boxes[box] = player
            break
        else:
            print('That box is already taken. Try another one.\n')

def get_computer_move():
    available_moves = [i for i, box in enumerate(boxes) if box == ' ']
    return random.choice(available_moves)

def switch_player(turn):
    return COMPUTER if turn % 2 == 0 else HUMAN

def check_for_win(player, turn):
    if turn >= 5:
        for combo in winning_combos:
            if all(boxes[i] == player for i in combo):
                return 'win'
    if turn == 9:
        return 'tie'
    return False

def play(player, turn):
    while True:
        take_turn(player, turn)
        print_board()
        result = check_for_win(player, turn)
        if result == 'win':
            print(f'Game over. {player} wins!\n')
            break
        elif result == 'tie':
            print("Game over. It's a tie!\n")
            break
        turn += 1
        player = switch_player(turn)

# Start the game
print('\nWelcome to Tic Tac Toe!')
print_board(initial=True)
play(first_player, turn)
```
